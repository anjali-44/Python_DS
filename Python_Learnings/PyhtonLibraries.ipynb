### Merging of two dataframes

```python
temp_df = pd.DataFrame ({
    "city" : ["chicago","new york","orlando"],
    "temperature" : [32,45,30],
    
})
temp_df
```

```python
wind_df = pd.DataFrame ({
    "city" : ["chicago","new york", "orlando"],
    "humidity" : [65,68,75]
})
```

```python
wind_df
```

```python
## merging on the basis of same city column
```

```python
df = pd.merge(temp_df, wind_df, on="city")
```

```python
df
```

```python
temp_df = pd.DataFrame ({
    "city" : ["chicago","new york","orlando","san francisco"],
    "temperature" : [32,45,30,40],
    
})
temp_df
```
